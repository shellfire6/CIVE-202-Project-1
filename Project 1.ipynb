```python
import pandas as pd
air_data = pd.read_csv('AirQuality_Daily_StudentVersion.csv')
ad = pd.DataFrame(air_data)
```

```python
advoc = ad.groupby(['sensor.name']).agg(vocmean = ('voc','mean'),
                                       vocmedian = ('voc','median'))

vocmean = advoc.sort_values(by = 'vocmean', ascending = False)
print(vocmean.head(5))

vocmedian = advoc.sort_values(by = 'vocmedian', ascending = False)
print(vocmedian.head(5))

ad25 = ad.groupby(['sensor.name']).agg(pmtwomean = ('pm2.5_atm','mean'),
                                       pmtwomedian = ('pm2.5_atm','median'))

pmtwomean = ad25.sort_values(by = 'pmtwomean', ascending = False)
print(pmtwomean.head(5))

pmtwomedian = ad25.sort_values(by = 'pmtwomedian', ascending = False)
print(pmtwomedian.head(5))

ad10 = ad.groupby(['sensor.name']).agg(pmtenmean = ('pm10.0_atm','mean'),
                                       pmtenmedian = ('pm10.0_atm','median'))

pmtenmean = ad10.sort_values(by = 'pmtenmean', ascending = False)
print(pmtenmean.head(5))

pmtenmedian = ad10.sort_values(by = 'pmtenmedian', ascending = False)
print(pmtenmedian.head(5))
```

```python
badvoc = ad.groupby(['date','sensor.name']).agg(vocmax = ('voc','max'))
worstvoc = badvoc.sort_values(by = 'vocmax', ascending = False)
print(worstvoc.head(1))

badpmtwo = ad.groupby(['date','sensor.name']).agg(pmtwomax = ('pm2.5_atm','max'))
worstpmtwo = badpmtwo.sort_values(by = 'pmtwomax', ascending = False)
print(worstpmtwo.head(1))

badpmten = ad.groupby(['date','sensor.name']).agg(pmtenmax = ('pm10.0_atm','max'))
worstpmten = badpmten.sort_values(by = 'pmtenmax', ascending = False)
print(worstpmten.head(1))
```

```python
def temp(x): 
    if x < 32:
        return 'Below Freezing'
    elif 32 <= x <= 50:
        return 'Cool'
    elif 50 < x <= 70:
        return 'Warm'
    else:
        return 'Hot'

def humidity(x): 
    if x < 50:
        return 'Low Humidity'
    elif 50 <= x <= 80:
        return 'High Humidity'
    else:
        return 'Very High Humidity'

ad['temperature_level'] = ad['temperature'].apply(temp)
ad['humidity_level'] = ad['humidity'].apply(humidity)



cold = ad[ad['temperature_level']=='Below Freezing']
cool = ad[ad['temperature_level']=='Cool']
warm = ad[ad['temperature_level']=='Warm']
hot = ad[ad['temperature_level']=='Hot']

low = ad[ad['humidity_level']=='Low Humidity']
high = ad[ad['humidity_level']=='High Humidity']
very = ad[ad['humidity_level']=='Very High Humidity']


tempvoc1 = cold.agg(cold_mean = ('voc','mean'))
print(tempvoc1)

tempvoc2 = cool.agg(cool_mean = ('voc','mean'))
print(tempvoc2)

tempvoc3 = warm.agg(warm_mean = ('voc','mean'))
print(tempvoc3)

tempvoc4 = hot.agg(hot_mean = ('voc','mean'))
print(tempvoc4)

temppm21 = cold.agg(cold_mean = ('pm2.5_atm','mean'))
print(temppm21)

temppm22 = cool.agg(cool_mean = ('pm2.5_atm','mean'))
print(temppm22)

temppm23 = warm.agg(warm_mean = ('pm2.5_atm','mean'))
print(temppm23)

temppm24 = hot.agg(hot_mean = ('pm2.5_atm','mean'))
print(temppm24)

temppm101 = cold.agg(cold_mean = ('pm10.0_atm','mean'))
print(temppm101)

temppm102 = cool.agg(cool_mean = ('pm10.0_atm','mean'))
print(temppm102)

temppm103 = warm.agg(warm_mean = ('pm10.0_atm','mean'))
print(temppm103)

temppm104 = hot.agg(hot_mean = ('pm10.0_atm','mean'))
print(temppm104)



humvoc1 = low.agg(low_mean = ('voc','mean'))
print(humvoc1)

humvoc2 = high.agg(high_mean = ('voc','mean'))
print(humvoc2)

humvoc3 = very.agg(very_high_mean = ('voc','mean'))
print(humvoc3)

humpm21 = low.agg(low_mean = ('pm2.5_atm','mean'))
print(humpm21)

humpm22 = high.agg(high_mean = ('pm2.5_atm','mean'))
print(humpm22)

humpm23 = very.agg(very_high_mean = ('pm2.5_atm','mean'))
print(humpm23)

humpm101 = low.agg(low_mean = ('pm10.0_atm','mean'))
print(humpm101)

humpm102 = high.agg(high_mean = ('pm10.0_atm','mean'))
print(humpm102)

humpm103 = very.agg(very_high_mean = ('pm10.0_atm','mean'))
print(humpm103)
```
